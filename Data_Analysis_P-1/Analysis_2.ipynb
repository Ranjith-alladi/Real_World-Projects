###  What is the Average price for vehicles based on the type of Vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis

```python
average_prices = df.groupby(['vehicleType','gearbox'])['price'].mean()
average_prices = pd.DataFrame(average_prices).reset_index()
average_price = average_prices[average_prices["vehicleType"] != "andere"]
average_price
```

```python
sns.barplot(x=average_price['vehicleType'],y=average_price['price'],hue=average_price['gearbox'])
plt.title('The Average price for vehicles based on the type of Vehicle as well as on the type of gearbox')
plt.xticks(rotation=45)
plt.show()
```

```python
average_price_1 = average_prices[average_prices["vehicleType"] == "andere"]
```

```python
average_price_1
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x=average_price_1['vehicleType'],y=average_price_1['price'],hue=average_price_1['gearbox'])
plt.title('The Average price for vehicle based on the type of Vehicle and gearbox for andere vehicletype')
plt.show()
```

```python
pivot_table = average_price.pivot(index="vehicleType", columns="gearbox", values="price")
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".2f")
```
